```python
np_cli_np_df = pd.merge(
    np_cli_df, np_df, on='ponuda_id', how='inner', suffixes=('_cli', '_np'))
```

```python
first_columns = ['klijent_id', 'sif_vrsta']
other_columns = [k for k in np_cli_np_df.columns if k not in first_columns]
new_order = first_columns + other_columns
np_cli_np_df = np_cli_np_df[new_order]
```

🧠 1. Šta radi groupby(...).head()
np_cli_np_df.groupby(['klijent_id', 'sif_vrsta']).head()

👉 Ovo znači:

grupiši po (klijent_id, sif_vrsta)
uzmi prvi red iz svake grupe

Zato vidiš normalne podatke.

🧠 2. Šta radi groupby(...).shift(1)
np_cli_np_df.groupby(['klijent_id', 'sif_vrsta']).shift(1)

👉 Ovo znači:
Za SVAKU grupu posebno:

pomjeri redove za 1 unazad
tj. svaki red dobija vrijednosti prethodnog reda iz iste grupe

Aaa kapiram da imaju 2 reda za grupu klijent_id sif_vrsta vrijednosti nebi bile null?

TAKO JE 👏🔥

Sad si ga pogodio 100%.

.expanding()

To znači:

👉 "uzmi sve od početka do trenutnog reda"

Bez ekspanding je samo prosli red a sa expanding je sve do proslog reda, a prosli red je shift(1)
