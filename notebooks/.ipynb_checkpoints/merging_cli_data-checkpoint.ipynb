```python
import pandas as pd

def prepare_cli_df(row):
    
    if pd.notna(row['osig_jmbg']):

        jmbg = str(row['osig_jmbg']).split('.')[0]

        day_month = jmbg[:4]
        year_part = int(jmbg[4:7])

        if year_part < 100:
            year = 2000 + year_part
        else:
            year = 1000 + year_part

        date_str = f"{day_month}{year}"

        date = pd.to_datetime(date_str, format='%d%m%Y', errors='coerce')

        print(date)
        print(date_str)
        return date

    return None
```

```python
cli_df.iloc[:4].apply(lambda x: prepare_cli_df(x), axis=1)
```
