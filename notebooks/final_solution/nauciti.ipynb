<h1>ML Inženjer – Pitanja i Odgovori</h1>

<h2>1. Kako bi napravio train/validation split za podatke od januara do decembra 2025. ako predviđaš januar 2026?</h2>

<p><strong>Odgovor:</strong></p>

<p>
Validation skup mora predstavljati budućnost u odnosu na train skup.
Primjer:
</p>

<ul>
    <li>Train: Januar – Oktobar 2025</li>
    <li>Validation: Novembar – Decembar 2025</li>
</ul>

<p>
Još bolje je koristiti rolling ili walk-forward validaciju:
</p>

<ul>
    <li>Train Jan–Jun → Validation Jul</li>
    <li>Train Jan–Jul → Validation Avg</li>
    <li>itd.</li>
</ul>

<p>
Cilj je da validacija što vjernije simulira produkciju.
</p>

<hr>

<h2>2. Model ima Train HR@5 = 0.91 i Validation HR@5 = 0.63. Šta prvo provjeravaš?</h2>

<p><strong>Odgovor:</strong></p>

<ol>
    <li>Da li postoji overfitting (velika varijansa).</li>
    <li>Da li postoji data leakage.</li>
    <li>Da li se distribucije train i validation podataka značajno razlikuju.</li>
</ol>

<hr>

<h2>3. Kako može nastati leakage kod feature-a "broj kupovina u posljednjih 30 dana"?</h2>

<p><strong>Odgovor:</strong></p>

<p>
Ako za događaj od 15.10.2025. računaš kupovine do 31.10.2025.,
koristiš podatke iz budućnosti.
</p>

<p>
Model tada vidi informacije koje u realnoj produkciji neće imati.
</p>

<hr>

<h2>4. Zašto random split nije dobar za vremenske podatke?</h2>

<p><strong>Odgovor:</strong></p>

<p>
Kod random splita budući podaci mogu završiti u train skupu.
</p>

<p>
Model tada indirektno uči iz budućnosti, pa validacioni rezultat izgleda
bolje nego što će biti u produkciji.
</p>

<h2>5. Objasni razliku između klasifikacije, regresije i rankinga.</h2>

<p><strong>Odgovor:</strong></p>

<h3>Klasifikacija</h3>

<p>
Predviđa kategoriju.
Primjer: Hoće li korisnik kupiti proizvod?
</p>

<h3>Regresija</h3>

<p>
Predviđa numeričku vrijednost.
Primjer: Koliko će korisnik potrošiti?
</p>

<h3>Ranking</h3>

<p>
Predviđa redoslijed rezultata.
Primjer: Kojih 10 proizvoda treba prikazati korisniku?
</p>

<p>
Za preporuke je ranking često prirodniji problem od klasifikacije.
</p>

<hr>

<h2>6. Objasni razliku između HR@5, MRR@5 i NDCG@5.</h2>

<p><strong>Odgovor:</strong></p>

<h3>HR@5</h3>

<p>
Provjerava da li se tačan rezultat nalazi među prvih 5.
</p>

<h3>MRR@5</h3>

<p>
Nagrađuje više pozicije.
Rezultat na prvom mjestu vrijedi više nego rezultat na petom.
</p>

<h3>NDCG@5</h3>

<p>
Ocjenjuje kvalitet kompletnog redoslijeda relevantnih rezultata.
</p>

<hr>

<h2>7. Da li je moguće da HR@5 raste, a MRR@5 pada?</h2>

<p><strong>Odgovor:</strong></p>

<p>Da.</p>

<p>
Ako model i dalje pogađa rezultat unutar top 5,
ali ga pomjera sa pozicije 1 na poziciju 5,
HR ostaje isti ili raste,
dok MRR opada.
</p>

<hr>

<h2>8. Zašto LightGBM Ranker koristi group?</h2>

<p><strong>Odgovor:</strong></p>

<p>
Ranking se radi unutar grupa.
</p>

<p>Primjer:</p>

<ul>
    <li>Korisnik A ima 10 kandidata.</li>
    <li>Korisnik B ima 20 kandidata.</li>
</ul>

<p>
To su dva odvojena ranking problema.
</p>

<p>
Ako group nije ispravno definisan,
model uči pogrešan ranking zadatak.
</p>

<h1>Gradient Boosting Stabla + Optuna (Dubinsko objašnjenje)</h1>

<p>
Ovaj dokument objašnjava kako boosting stabla rade ispod haube, kako se treniraju, 
koji su ključni parametri i kako optimizacija hiperparametara (Optuna) zapravo funkcioniše.
</p>

<hr>

<h2>1. Šta je Gradient Boosting?</h2>

<p>
Gradient Boosting je metoda u kojoj se veliki model gradi kao suma mnogo malih modela (stabala).
Svako novo stablo ne uči direktno target, nego uči <b>grešku prethodnog modela</b>.
</p>

<p><b>Osnovna ideja:</b></p>

<blockquote>
Model se gradi iterativno tako da svako novo stablo smanjuje grešku prethodnog modela.
</blockquote>

<hr>

<h2>2. Intuicija iza Boosting-a</h2>

<p>Zamisli da pokušavaš predvidjeti cijene stanova:</p>

<ul>
    <li>Prvo napraviš vrlo jednostavan model (npr. prosjek cijene).</li>
    <li>Onda vidiš gdje griješi.</li>
    <li>Sljedeće stablo uči te greške.</li>
    <li>Svako novo stablo popravlja prethodni model.</li>
</ul>

<p>
Na kraju, imaš mnogo malih “ispravljača grešaka”.
</p>

<hr>

<h2>3. Matematička ideja (ključna stvar)</h2>

<p>
Model se gradi kao suma:
</p>

<p>
<b>F(x) = F₀(x) + η * T₁(x) + η * T₂(x) + ... + η * Tₙ(x)</b>
</p>

<ul>
    <li>F(x) → finalni model</li>
    <li>F₀(x) → početna predikcija (npr. prosjek)</li>
    <li>Tᵢ(x) → i-to stablo</li>
    <li>η → learning rate</li>
</ul>

<hr>

<h2>4. Kako se trenira jedno stablo u boosting-u?</h2>

<p>Proces ide ovako:</p>

<h3>Korak 1: početna predikcija</h3>

<p>
Model kreće sa jednostavnom predikcijom:
</p>

<blockquote>
ŷ₀ = prosjek targeta
</blockquote>

<h3>Korak 2: računanje greške</h3>

<p>
Greška (residual):
</p>

<blockquote>
r = y - ŷ
</blockquote>

<p>
U naprednijem obliku koristi se gradient loss funkcije:
</p>

<blockquote>
r = -∂L/∂ŷ
</blockquote>

<h3>Korak 3: treniranje stabla</h3>

<p>
Novo stablo uči da predviđa greške prethodnog modela.
</p>

<h3>Korak 4: update modela</h3>

<blockquote>
F(x) = F(x) + η * T(x)
</blockquote>

<h3>Korak 5: ponavljanje</h3>

<p>
Proces se ponavlja N puta (n_estimators).
</p>

<hr>

<h2>5. Zašto se zove “Gradient” Boosting?</h2>

<p>
Zato što se optimizacija zasniva na <b>gradijentu loss funkcije</b>.
</p>

<p>
Model se ponaša kao da radi gradient descent, ali umjesto brojeva koristi stabla.
</p>

<hr>

<h2>6. Bias-Variance kod Boosting-a</h2>

<ul>
    <li><b>Bias</b> → premalo kompleksan model (underfitting)</li>
    <li><b>Variance</b> → previše osjetljiv model (overfitting)</li>
</ul>

<p>
Boosting:
</p>

<ul>
    <li>postepeno smanjuje bias</li>
    <li>može povećati variance ako nije kontrolisan</li>
</ul>

<hr>

<h1>7. Ključni parametri Boosting modela</h1>

<hr>

<h2>7.1 learning_rate</h2>

<p>
Kontroliše koliko jako svako novo stablo utiče na model.
</p>

<blockquote>
F(x) = F(x) + η * T(x)
</blockquote>

<ul>
    <li>Velik learning rate → brzo učenje, veći rizik overfitting-a</li>
    <li>Mali learning rate → stabilno, ali zahtijeva više stabala</li>
</ul>

<hr>

<h2>7.2 n_estimators</h2>

<p>
Broj stabala u modelu.
</p>

<ul>
    <li>Premalo → underfitting</li>
    <li>Previše → može overfitting (ako nema regularizacije)</li>
</ul>

<hr>

<h2>7.3 max_depth</h2>

<p>
Maksimalna dubina stabla.
</p>

<ul>
    <li>Veća dubina → kompleksniji model</li>
    <li>Manja dubina → jednostavniji model</li>
</ul>

<p>
Duboka stabla imaju veću varijansu.
</p>

<hr>

<h2>7.4 num_leaves (LightGBM)</h2>

<p>
Broj listova u stablu.
</p>

<p>
Svaki leaf je jedna konačna predikcija.
</p>

<blockquote>
Više listova = veća fleksibilnost = veći rizik overfitting-a
</blockquote>

<hr>

<h2>7.5 min_data_in_leaf</h2>

<p>
Minimalan broj uzoraka u jednom listu.
</p>

<ul>
    <li>Povećava stabilnost</li>
    <li>Smanjuje overfitting</li>
</ul>

<hr>

<h2>7.6 feature_fraction</h2>

<p>
Procenat feature-a koji se koristi po stablu.
</p>

<ul>
    <li>Smanjuje korelaciju stabala</li>
    <li>Dodaje randomizaciju</li>
</ul>

<hr>

<h2>7.7 bagging_fraction</h2>

<p>
Procenat podataka koji se koristi po stablu.
</p>

<ul>
    <li>Smanjuje variance</li>
    <li>Dodaje robustnost</li>
</ul>

<hr>

<h2>7.8 lambda_l1 i lambda_l2</h2>

<ul>
    <li><b>L1</b> → gura feature težine na nulu (feature selection)</li>
    <li><b>L2</b> → smanjuje ekstremne vrijednosti</li>
</ul>

<h1>8. Optuna – optimizacija hiperparametara</h1>

<hr>

<h2>8.1 Šta je Optuna?</h2>

<p>
:contentReference[oaicite:0]{index=0} je framework za automatsko traženje najboljih hiperparametara modela.
</p>

<hr>

<h2>8.2 Problem koji rješava</h2>

<p>
Svaki model ima ogroman broj kombinacija parametara:
</p>

<ul>
    <li>learning_rate</li>
    <li>depth</li>
    <li>num_leaves</li>
    <li>regularization</li>
</ul>

<p>
Cilj je pronaći najbolju kombinaciju bez testiranja svih opcija.
</p>

<hr>

<h2>8.3 Zašto nije grid search?</h2>

<p>
Grid search testira SVE kombinacije → sporo i neefikasno.
</p>

<p>
Optuna koristi pametno pretraživanje prostora.
</p>

<hr>

<h2>8.4 Kako Optuna radi (TPE)</h2>

<p>
Optuna koristi probabilistički model:
</p>

<ul>
    <li>dijeli eksperimente na “dobre” i “loše”</li>
    <li>uči distribuciju dobrih parametara</li>
    <li>predlaže nove bolje kombinacije</li>
</ul>

<hr>

<h2>8.5 Pruning (vrlo važno)</h2>

<p>
Ako model rano izgleda loše, Optuna prekida trening.
</p>

<blockquote>
Ne troši resurse na loše eksperimente.
</blockquote>

<hr>

<h2>8.6 Optuna workflow</h2>

<ol>
    <li>Definiše search space</li>
    <li>Trenira model</li>
    <li>Računa metric (npr. HR@5)</li>
    <li>Vraća rezultat Optuni</li>
    <li>Optuna uči iz rezultata</li>
</ol>

<hr>

<h2>9. Kako sve povezati u jednu sliku</h2>

<p><b>Boosting:</b></p>
<blockquote>
Iterativno smanjivanje greške kroz niz stabala
</blockquote>

<p><b>Parametri:</b></p>
<blockquote>
Kontrolišu balans između bias-a i variance-a
</blockquote>

<p><b>Optuna:</b></p>
<blockquote>
Pametno traženje najboljih parametara tog balansa
</blockquote>

<hr>

<h2>10. Ključna intuicija (najvažnije)</h2>

<p>
Ako razumiješ samo ovo:
</p>

<blockquote>
Boosting = niz korekcija grešaka<br>
Parametri = kontrola kompleksnosti i stabilnosti<br>
Optuna = automatsko podešavanje tog balansa
</blockquote>

<h1>Kako funkcioniše učenje u Boosting stablima (LightGBM / XGBoost)</h1>

<p>
Ovo objašnjenje prikazuje tačan proces kako se stabla u boosting modelima uče jedno za drugim,
i šta zapravo znači “učenje greške prethodnog stabla”.
</p>

<hr>

<h2>1. Šta je početak modela?</h2>

<p>
Boosting model ne počinje “pametno”, nego sa najjednostavnijom mogućom predikcijom.
</p>

<h3>Za regresiju:</h3>

<blockquote>
F₀(x) = mean(y)
</blockquote>

<h3>Intuicija:</h3>

<p>
Na početku model kaže:
<b>“Svim primjerima dajem istu vrijednost.”</b>
</p>

<hr>

<h2>2. Šta prvo stablo zapravo uči?</h2>

<p>
Nakon početne predikcije, izračunava se greška:
</p>

<blockquote>
error = y - F₀(x)
</blockquote>

<p><b>Ključna ideja:</b></p>

<blockquote>
Prvo stablo ne uči target (y), nego uči grešku početnog modela.
</blockquote>

<h3>Intuicija:</h3>

<p>
Ako model svima daje 0.5, a stvarne vrijednosti su 0.1 i 0.9,
prvo stablo uči gdje treba povećati, a gdje smanjiti predikciju.
</p>

<hr>

<h2>3. Kako izgleda prvo stablo?</h2>

<p>
Stablo je skup if-then pravila:
</p>

<pre>
if feature_1 < 10:
    +0.3
else:
    -0.2
</pre>

<p>
Ove vrijednosti nisu target vrijednosti,
nego <b>korekcije greške</b>.
</p>

<hr>

<h2>4. Kako izgleda drugo stablo?</h2>

<p>
Nakon prvog stabla, model postaje:
</p>

<blockquote>
F₁(x) = F₀(x) + η * T₁(x)
</blockquote>

<p>
Sada se ponovo računa greška:
</p>

<blockquote>
error₁ = y - F₁(x)
</blockquote>

<p>
Drugo stablo uči samo ono što prvi model nije uspio popraviti.
</p>

<h3>Intuicija:</h3>

<p>
Prvo stablo rješava “grube greške”,
drugo stablo rješava “preostale greške”.
</p>

<hr>

<h2>5. Kako izgleda treće, četvrto i N-to stablo?</h2>

<p>
Svako novo stablo radi istu stvar:
</p>

<ul>
    <li>vidi preostalu grešku</li>
    <li>uči kako da je smanji</li>
    <li>dodaje sitne korekcije</li>
</ul>

<p>
Kako model napreduje:
</p>

<ul>
    <li>greške postaju manje</li>
    <li>korekcije postaju finije</li>
</ul>

<hr>

<h2>6. Šta znači “učenje greške”?</h2>

<p>
Stablo ne uči direktno target.
</p>

<p>
Umjesto toga, uči:
</p>

<blockquote>
Kako da promijenim predikciju da bude bliže stvarnoj vrijednosti?
</blockquote>

<h3>Matematički oblik:</h3>

<blockquote>
r = -∂L / ∂F(x)
</blockquote>

<p>
To je gradijent loss funkcije.
</p>

<hr>

<h2>7. Šta drugo stablo zna o prvom?</h2>

<p>
Drugo stablo ne vidi prvo stablo direktno.
</p>

<p>
Ono vidi samo:
</p>

<blockquote>
šta je ostalo pogrešno nakon prvog modela
</blockquote>

<h3>Intuicija:</h3>

<ul>
    <li>Prvo stablo popravlja veliki dio greške</li>
    <li>Drugo stablo vidi samo preostali dio</li>
    <li>Treće stablo vidi još manji preostali dio</li>
</ul>

<hr>

<h2>8. Zašto boosting radi dobro?</h2>

<p>
Zato što svaki novi model radi na jednostavnijem problemu.
</p>

<ul>
    <li>Prvo stablo → gruba aproksimacija</li>
    <li>Sljedeća stabla → postepene korekcije</li>
    <li>Zadnja stabla → fine optimizacije</li>
</ul>

<hr>

<h2>9. Intuicija u jednoj rečenici</h2>

<blockquote>
Svako stablo u boosting modelu uči greške prethodnog modela, a ne originalni problem.
</blockquote>

<h2>10. Mentalna slika</h2>

<p>
Zamisli model kao penjanje na planinu:
</p>

<ul>
    <li>F₀ → početna ravna linija</li>
    <li>T₁ → veliki korak ka vrhu</li>
    <li>T₂ → korekcija smjera</li>
    <li>T₃ → fine korekcije</li>
</ul>

<blockquote>
Svako stablo pomjera model bliže pravoj vrijednosti.
