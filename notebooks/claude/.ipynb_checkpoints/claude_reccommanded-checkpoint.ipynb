```python
"""
Pipeline za preporuku osiguranja klijentima
==========================================
Kombinovani pristup: Collaborative Filtering + Content-Based

Struktura ulaznih podataka:
- df_polise: podaci o polisama (ponuda_id, polisa_id, sif_vrsta, konacna_premija, ...)
- df_klijenti: podaci o klijentima (ponuda_id, osig_jmbg, pol_id, sif_delatnost, ...)
"""

import pandas as pd
import numpy as np
import hashlib
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import precision_score, recall_score, roc_auc_score
import warnings
warnings.filterwarnings('ignore')


# ===========================================================================
# KORAK 0: Učitavanje podataka
# ===========================================================================

def load_data(path_polise: str, path_klijenti: str):
    """Učitaj oba CSV fajla."""
    df_polise = pd.read_csv(path_polise)
    df_klijenti = pd.read_csv(path_klijenti)
    return df_polise, df_klijenti


# ===========================================================================
# KORAK 1: Priprema i čišćenje podataka
# ===========================================================================

def prepare_data(df_polise: pd.DataFrame, df_klijenti: pd.DataFrame) -> pd.DataFrame:
    """
    Spoji polise i klijente, izgradi osnovni dataset po klijentu.
    Vraća jedan red po (klijent × polisa) kombinaciji.
    """
    # 1.1 Anonimizacija JMBG — nikad ne koristimo pravi JMBG u modelu
    df_klijenti = df_klijenti.copy()
    df_klijenti['klijent_id'] = df_klijenti['osig_jmbg'].astype(str).apply(
        lambda x: hashlib.sha256(x.encode()).hexdigest()[:12]
    )

    # 1.2 Izvedi starost iz JMBG (prvih 7 cifara = DDMMGGG)
    def jmbg_to_age(jmbg):
        try:
            jmbg = str(jmbg).zfill(13)
            dan = int(jmbg[0:2])
            mj  = int(jmbg[2:4])
            god = int(jmbg[4:7])
            # JMBG: 000-899 = 2000+, 900-999 = 1900+
            god = (2000 + god) if god < 900 else (1000 + god)
            from datetime import date
            return date.today().year - god
        except:
            return np.nan

    df_klijenti['starost'] = df_klijenti['osig_jmbg'].apply(jmbg_to_age)

    # 1.3 Uzmi samo jednu pojavu po klijentu za demografske podatke
    klijent_profil = df_klijenti.drop_duplicates(subset='klijent_id')[[
        'ponuda_id', 'klijent_id', 'pol_id', 'starost',
        'sif_delatnost', 'osig_mesto'
    ]]

    # klijent_profil = df_klijenti.groupby('klijent_id').agg({
    #     'ponuda_id': 'max',              # npr. najnovija
    #     'pol_id': lambda x: x.dropna().iloc[0] if not x.dropna().empty else np.nan,
    #     'starost': 'max',                # ili mean
    #     'sif_delatnost': lambda x: x.dropna().iloc[0] if not x.dropna().empty else np.nan,
    #     'osig_mesto': lambda x: x.dropna().iloc[0] if not x.dropna().empty else np.nan,
    # }).reset_index()
    
    # 1.4 Spoji polise sa profilom klijenta
    df = df_polise.merge(
        df_klijenti[['ponuda_id', 'klijent_id']].drop_duplicates(),
        on='ponuda_id',
        how='left'
    )

    # 1.5 Ukloni redove bez sif_vrsta ili klijent_id
    df = df.dropna(subset=['sif_vrsta', 'klijent_id'])
    df = df[df['sif_vrsta'] != 0]  # vrsta 0 su prazne/nedefinirane

    print(f"✓ Ukupno redova nakon čišćenja: {len(df)}")
    print(f"✓ Jedinstvenih klijenata: {df['klijent_id'].nunique()}")
    print(f"✓ Vrsta polisa: {df['sif_vrsta'].nunique()}")
    print(f"  Vrste: {sorted(df['sif_vrsta'].unique().tolist())}")

    return df, klijent_profil


# ===========================================================================
# KORAK 2: Izgradnja klijent × polisa matrice
# ===========================================================================

def build_interaction_matrix(df: pd.DataFrame) -> pd.DataFrame:
    """
    Gradi binarnu matricu: redovi = klijenti, kolone = vrste polisa.
    Vrijednost 1 = klijent ima tu polisu, 0 = nema.
    """
    matrix = df.groupby(['klijent_id', 'sif_vrsta']).size().unstack(fill_value=0)
    # Konvertuj u binarno (ne zanima nas koliko puta, nego ima li ili ne)
    matrix = (matrix > 0).astype(int)

    print(f"\n✓ Matrica interakcija: {matrix.shape[0]} klijenata × {matrix.shape[1]} vrsta polisa")
    print(f"  Gustoća matrice: {matrix.values.mean():.2%} (% popunjenih ćelija)")

    return matrix


# ===========================================================================
# KORAK 3: Collaborative Filtering (User-Based)
# ===========================================================================

class CollaborativeFilteringModel:
    """
    User-based collaborative filtering koristeći cosine similarity.
    Preporučuje polise na osnovu sličnih klijenata.
    """

    def __init__(self, top_n_similar: int = 50):
        self.top_n_similar = top_n_similar
        self.matrix = None
        self.similarity_matrix = None

    def fit(self, matrix: pd.DataFrame):
        self.matrix = matrix
        print("\n⏳ Računam sličnost između klijenata...")
        self.similarity_matrix = pd.DataFrame(
            cosine_similarity(matrix.values),
            index=matrix.index,
            columns=matrix.index
        )
        print("✓ Similarity matrica izgrađena.")
        return self

    def recommend(self, klijent_id: str, top_n: int = 5) -> list:
        """
        Za datog klijenta vrati top_n preporučenih vrsta polisa
        koje klijent još nema.
        """
        if klijent_id not in self.matrix.index:
            return []

        # Pronađi top N najsličnijih klijenata
        sim_scores = self.similarity_matrix[klijent_id].drop(klijent_id).nlargest(self.top_n_similar)
        similar_clients = sim_scores.index.tolist()

        # Ponderirani zbir polisa sličnih klijenata
        weighted_scores = self.matrix.loc[similar_clients].T.dot(sim_scores)

        # Ukloni polise koje klijent već ima
        already_has = self.matrix.loc[klijent_id]
        weighted_scores = weighted_scores[already_has == 0]

        # Top N preporuka
        recommendations = weighted_scores.nlargest(top_n).index.tolist()
        return recommendations

    def recommend_batch(self, top_n: int = 5) -> pd.DataFrame:
        """Generiši preporuke za sve klijente."""
        results = []
        for klijent_id in self.matrix.index:
            recs = self.recommend(klijent_id, top_n)
            for rank, vrsta in enumerate(recs, 1):
                results.append({
                    'klijent_id': klijent_id,
                    'preporucena_vrsta': vrsta,
                    'rank': rank
                })
        return pd.DataFrame(results)


# ===========================================================================
# KORAK 4: Content-Based Model (za cold start)
# ===========================================================================

class ContentBasedModel:
    """
    Za svaку vrstu polise trenira binarni klasifikator.
    Koristi demografske feature-e klijenta.
    Korisno za nove klijente bez historije polisa.
    """

    def __init__(self):
        self.models = {}
        self.feature_cols = ['pol_id', 'starost', 'sif_delatnost']
        self.label_encoders = {}

    def _prepare_features(self, df_klijenti_profil: pd.DataFrame,
                          matrix: pd.DataFrame) -> pd.DataFrame:
        """Spoji profile sa matricom, pripremi feature-e."""
        data = matrix.reset_index().merge(
            df_klijenti_profil[['klijent_id'] + self.feature_cols],
            on='klijent_id',
            how='left'
        )

        # Enkoduj kategoričke varijable
        for col in ['pol_id', 'sif_delatnost']:
            if col in data.columns:
                le = LabelEncoder()
                data[col] = data[col].fillna('NEPOZNATO').astype(str)
                data[col] = le.fit_transform(data[col])
                self.label_encoders[col] = le

        data['starost'] = data['starost'].fillna(data['starost'].median())
        return data

    def fit(self, df_klijenti_profil: pd.DataFrame, matrix: pd.DataFrame):
        """Treniraj model za svaku vrstu polise."""
        data = self._prepare_features(df_klijenti_profil, matrix)
        X = data[self.feature_cols].values
        vrste_polisa = matrix.columns.tolist()

        print(f"\n⏳ Treniram content-based modele za {len(vrste_polisa)} vrsta polisa...")

        for vrsta in vrste_polisa:
            y = data[vrsta].values
            if y.sum() < 10:  # Preskoči ako premalo primjera
                continue
            clf = RandomForestClassifier(n_estimators=100, random_state=42,
                                          class_weight='balanced')
            clf.fit(X, y)
            self.models[vrsta] = clf

        print(f"✓ Trenirano {len(self.models)} modela.")
        return self

    def recommend_new_client(self, profil: dict, top_n: int = 5) -> list:
        """
        Za novog klijenta (bez historije) generiši preporuke.
        profil = {'pol_id': 2, 'starost': 45, 'sif_delatnost': None}
        """
        scores = {}
        for vrsta, model in self.models.items():
            row = [
                profil.get('pol_id', 0),
                profil.get('starost', 40),
                profil.get('sif_delatnost', 0)
            ]
            prob = model.predict_proba([row])[0][1]
            scores[vrsta] = prob

        top = sorted(scores.items(), key=lambda x: x[1], reverse=True)[:top_n]
        return [(vrsta, round(prob, 3)) for vrsta, prob in top]


# ===========================================================================
# KORAK 5: Evaluacija
# ===========================================================================

def evaluate_collaborative(model: CollaborativeFilteringModel,
                             test_interactions: pd.DataFrame,
                             top_n: int = 5) -> dict:
    """
    Leave-one-out evaluacija:
    Za svakog klijenta sakrij jednu polisu, provjeri da li je model preporuči.
    Vraća Precision@K i Recall@K.
    """
    hits = 0
    total = 0

    for klijent_id in test_interactions.index:
        aktuelne = test_interactions.loc[klijent_id]
        polise_klijenta = aktuelne[aktuelne == 1].index.tolist()

        if len(polise_klijenta) < 2:
            continue

        # Sakrij jednu polisu
        sakrivena = polise_klijenta[-1]
        preporuke = model.recommend(klijent_id, top_n=top_n)

        if sakrivena in preporuke:
            hits += 1
        total += 1

    precision_at_k = hits / total if total > 0 else 0
    print(f"\n📊 Evaluacija Collaborative Filtering (top {top_n}):")
    print(f"   Hit Rate (Precision@{top_n}): {precision_at_k:.2%}")
    print(f"   Testirano klijenata: {total}")

    return {'hit_rate': precision_at_k, 'tested_clients': total}


# ===========================================================================
# KORAK 6: Glavni pipeline
# ===========================================================================

def run_pipeline(path_polise: str, path_klijenti: str):
    """Pokreni cijeli pipeline."""

    print("=" * 60)
    print("PIPELINE ZA PREPORUKU OSIGURANJA")
    print("=" * 60)

    # Učitaj podatke
    df_polise, df_klijenti = load_data(path_polise, path_klijenti)

    # Pripremi podatke
    df, klijent_profil = prepare_data(df_polise, df_klijenti)

    # Izgradi matricu
    matrix = build_interaction_matrix(df)

    # Podijeli na train/test (80/20 klijenata)
    train_idx, test_idx = train_test_split(matrix.index, test_size=0.2, random_state=42)
    train_matrix = matrix.loc[train_idx]
    test_matrix  = matrix.loc[test_idx]

    # Collaborative Filtering
    cf_model = CollaborativeFilteringModel(top_n_similar=50)
    cf_model.fit(train_matrix)

    # Evaluacija
    evaluate_collaborative(cf_model, test_matrix, top_n=5)

    # Content-Based (za cold start)
    cb_model = ContentBasedModel()
    cb_model.fit(klijent_profil, train_matrix)

    # Primjer preporuke za postojećeg klijenta
    primjer_klijent = train_matrix.index[0]
    print(f"\n🎯 Primjer preporuke za klijenta {primjer_klijent}:")
    print(f"   Ima polise: {train_matrix.loc[primjer_klijent][train_matrix.loc[primjer_klijent]==1].index.tolist()}")
    print(f"   Preporuke:  {cf_model.recommend(primjer_klijent, top_n=5)}")

    # Primjer za novog klijenta
    print(f"\n🆕 Primjer preporuke za novog klijenta (content-based):")
    novi_klijent = {'pol_id': 2, 'starost': 42, 'sif_delatnost': None}
    preporuke_novi = cb_model.recommend_new_client(novi_klijent, top_n=5)
    print(f"   Profil: {novi_klijent}")
    print(f"   Preporuke: {preporuke_novi}")

    # Generiši preporuke za sve klijente i sačuvaj
    print("\n⏳ Generišem preporuke za sve klijente...")
    sve_preporuke = cf_model.recommend_batch(top_n=5)
    sve_preporuke.to_csv('preporuke_output.csv', index=False)
    print(f"✓ Sačuvano {len(sve_preporuke)} preporuka u 'preporuke_output.csv'")

    return cf_model, cb_model, matrix


# ===========================================================================
# Pokretanje
# ===========================================================================

if __name__ == "__main__":
    # Zamijeni putanjama do tvojih CSV fajlova
    cf_model, cb_model, matrix = run_pipeline(
        path_polise="polise.csv",
        path_klijenti="klijenti.csv"
    )
```
